ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
# 2.Problemin Çözümünü Buraya Yazınız
def degisim():
    dosya_adi = input("dosay adi giriniz :")
    eski_isim= input("değişmek istediğiniz ismi giriniz :")
    yeni_isim = input("yeni ismi giriniz :")
    with open(dosya_adi, "r", encoding="utf-8") as file:
        bilgi = file.read()
    yeni_bilgi = bilgi.replace(eski_isim,yeni_isim)
    with open(dosya_adi, "w", encoding="utf-8") as dosya:
        dosya.write(yeni_bilgi)
degisim()
````
